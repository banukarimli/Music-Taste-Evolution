Read the CSV file and create the DataFrame:

```python
import pandas as pd

csv_file_path = '/Users/banukarimli/210 proj/playlist.csv'
json_file_path = '/Users/banukarimli/210 proj/StreamingHistory0.json'
df = pd.read_csv(csv_file_path)
df = df[df["#"] <= 20]
df.head()
```

```python
df.describe()
```

Display Top 10 Genres in the playlist based on their frequency:

```python
import matplotlib.pyplot as plt
import seaborn as sns

all_genres = df['Genres'].str.split(', ').explode()
top_10_genres = all_genres.value_counts().head(10)

plt.figure(figsize=(12, 6))
sns.barplot(x=top_10_genres.values, y=top_10_genres.index, hue=top_10_genres.index, palette='viridis', legend=False)
plt.title('Top 10 Genres')
plt.xlabel('Frequency')
plt.ylabel('Genre')
plt.show()
```

Display Top 10 Artists:

```python
top_artists = df['Artist'].value_counts().head(10)
print("Top 10 Artists 🎶")
print("-----------------")
for rank, (artist, count) in enumerate(top_artists.items(), start=1):
    print(f"{rank}. {artist} - {count} songs")
```

Danceability of Top 10 Genres:

```python
top_10_genres = df['Genres'].value_counts().head(10)
sns.set(style='darkgrid')

plt.figure(figsize=(14, 8))
sns.boxplot(x='Dance', y='Genres', data=df[df['Genres'].isin(top_10_genres.index)], palette='muted')
plt.title('Danceability of Top 10 Genres')
plt.xlabel('Danceability')
plt.ylabel('Genres')
plt.show()
```

```python
df['Added At'] = pd.to_datetime(df['Added At'])

top_genres = df['Genres'].value_counts().head(5).index

df_top_genres = df[df['Genres'].isin(top_genres)]
genre_counts_over_time = df_top_genres.groupby(['Added At', 'Genres']).size().unstack(fill_value=0)

sns.set(style='darkgrid')
plt.figure(figsize=(16, 8))
genre_counts_over_time.plot(kind='line', marker='o', colormap='tab10', lw=2)
plt.title('Top Genres Over Time')
plt.xlabel('Date Added to Playlist')
plt.ylabel('Number of Songs')
plt.legend(title='Genres', bbox_to_anchor=(1, 1))
plt.show()
```

```python
numerical_columns = df.select_dtypes(include=['float64', 'int64'])
correlation_matrix = numerical_columns.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Matrix')
plt.show()
```

```python
df['Added At'] = pd.to_datetime(df['Added At'])
df['Period'] = df['Added At'].dt.to_period('M')
genres_counts = df.groupby(['Period', 'Genres']).size().reset_index(name='Count')

genres_pivot = genres_counts.pivot(index='Period', columns='Genres', values='Count').fillna(0)

periods = genres_pivot.index
num_periods = len(periods)

plt.figure(figsize=(15, 5 * num_periods))

for i, period in enumerate(periods, start=1):
    plt.subplot(num_periods, 1, i)
    plt.title(f'Genres Distribution - {period}')
    plt.pie(genres_pivot.loc[period], labels=genres_pivot.columns, autopct='%1.1f%%', startangle=90)
    plt.axis('equal')

plt.tight_layout()
plt.show()
```

```python

df['Added At'] = pd.to_datetime(df['Added At'])
df = df.sort_values(by='Added At')

midpoint_index = len(df) // 2

df_beginning = df.iloc[:midpoint_index]
df_end = df.iloc[midpoint_index:]

top_genres_beginning = df_beginning['Genres'].value_counts().head(10)
top_genres_end = df_end['Genres'].value_counts().head(10)

plt.figure(figsize=(15, 10))
plt.subplot(1, 2, 1)
top_genres_beginning.plot(kind='bar', color='skyblue')
plt.title('Top Genres - Beginning')
plt.subplot(1, 2, 2)
top_genres_end.plot(kind='bar', color='skyblue')
plt.title('Top Genres - End')
plt.tight_layout()
plt.show()
```

```python
top_artists_beginning = df_beginning['Artist'].value_counts().head(10)
top_artists_end = df_end['Artist'].value_counts().head(10)

plt.figure(figsize=(15, 10))
plt.subplot(1, 2, 1)
top_artists_beginning.plot(kind='bar', color='lightcoral')
plt.title('Top Artists - Beginning')

plt.subplot(1, 2, 2)
top_artists_end.plot(kind='bar', color='lightcoral')
plt.title('Top Artists - End')

plt.tight_layout()
plt.show()
```
